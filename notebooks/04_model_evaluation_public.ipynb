```python
import joblib
import os

load_path = "/content/drive/MyDrive/Professional/Portfolio/Transformations/"

# Verify the file paths
missing_files = []

for file in ['X_train.pkl', 'X_test.pkl', 'y_train.pkl', 'y_test.pkl']:
    if not os.path.exists(load_path + file):
        print(f"‚ùå {file} not found. Please check the file path or regenerate the data.")
        missing_files.append(file)

if missing_files:
    print("\n‚ö†Ô∏è The following files are missing: " + ", ".join(missing_files))
    print("üö® Please regenerate missing files before proceeding.")
else:
    X_train = joblib.load(load_path + 'X_train.pkl')
    X_test = joblib.load(load_path + 'X_test.pkl')
    y_train = joblib.load(load_path + 'y_train.pkl')
    y_test = joblib.load(load_path + 'y_test.pkl')
    print("‚úÖ All required data files loaded successfully!")
```

# Interaction Terms

```python
interaction_features = ['RM', 'LSTAT']
for feature in interaction_features:
    X_train[f'{feature}_x_{feature}'] = X_train[feature] * X_train[feature]
    X_test[f'{feature}_x_{feature}'] = X_test[feature] * X_test[feature]
```

# Polynomial Features, Deg=2

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = pd.DataFrame(poly.fit_transform(X_train), columns=poly.get_feature_names_out(X_train.columns))
X_test_poly = pd.DataFrame(poly.transform(X_test), columns=poly.get_feature_names_out(X_test.columns))
```

# Bin

```python
# Binning 'AGE' feature
bins = [0, 18, 35, 60, np.inf]
labels = ['Child', 'Young Adult', 'Middle-Aged', 'Senior']
X_train['AGE_Bin'] = pd.cut(X_train['AGE'], bins=bins, labels=labels)
X_test['AGE_Bin'] = pd.cut(X_test['AGE'], bins=bins, labels=labels)

# Binning 'DIS' feature (Distance to employment centers)
bins_dis = [0, 2, 4, np.inf]
labels_dis = ['Near', 'Moderate', 'Far']
X_train['DIS_Bin'] = pd.cut(X_train['DIS'], bins=bins_dis, labels=labels_dis)
X_test['DIS_Bin'] = pd.cut(X_test['DIS'], bins=bins_dis, labels=labels_dis)

# Binning 'CRIM' feature (Crime rate)
bins_crim = [0, 1, 5, np.inf]
labels_crim = ['Low', 'Medium', 'High']
X_train['CRIM_Bin'] = pd.cut(X_train['CRIM'], bins=bins_crim, labels=labels_crim)
X_test['CRIM_Bin'] = pd.cut(X_test['CRIM'], bins=bins_crim, labels=labels_crim)
```

# Save Data
